## Analysis of Nigeria's Food Trade (1995 - 2023)

#### Data Source: https://unctadstat.unctad.org/datacentre/dataviewer/US.TradeFoodCatByProc

### Data Importing, Cleaning, & Transformation

```python
# importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
```

```python
#importing datasets

#dataset of Nigeria's Food export monetary value
df_1 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProc_Export_$USMillions.csv") 

#dataset of Nigeria's Food export showing % of total food export for different food categories and process
df_2 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExport_%totalfood.csv") 

#dataset of Nigeria's Food export as percentage of total merchandise export
df_3 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExport_%totalMerchandise.csv")

#dataset of Nigeria's Food export year on year growth rate
df_4 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExportYoYGR.csv")

#dataset of Nigeria's Food import monetary value
df_5 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport$USMilliion.csv")

#dataset of Nigeria's Food import showing % of total food export for different food categories and process
df_6 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport%TotalFood.csv")

#dataset of Nigeria's Food import as percentage of total merchandise export
df_7 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport%TotalMerchandise.csv")

#dataset of Nigeria's Food import year on year growth rate
df_8 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcimportYoYGR.csv")
```

```python
# dropping & renaming columns
df_1 = df_1.drop(columns = ['US_at_current_prices_Footnote','US_at_current_prices_MissingValue' ])
df_1 = df_1.rename(columns = {'US_at_current_prices_in_millions_Value':'Export_$USMillion'})
df_1
```

```python
# dropping & renaming columns
df_2 = df_2.drop(columns = ['Percentage_of_total_food_Footnote','Percentage_of_total_food_MissingValue'])
df_2 = df_2.rename(columns = {'Percentage_of_total_food_Value':'Export_%TotalFood'})
df_2
```

```python
# dropping & renaming columns
df_3 = df_3.drop(columns = ['Percentage_of_total_merchandise_trade_Footnote','Percentage_of_total_merchandise_trade_MissingValue'])
df_3 = df_3.rename(columns = {'Percentage_of_total_merchandise_trade_Value':'Export_%TotalMerchandise'})
df_3
```

```python
# dropping & renaming columns
df_4 = df_4.drop(columns = ['Growth_rate_yearonyear_Footnote','Growth_rate_yearonyear_MissingValue'])
df_4 = df_4.rename(columns = {'Growth_rate_yearonyear_Value':'Export_YoYGR'})
df_4
```

```python
# dropping & renaming columns
df_5 = df_5.drop(columns = ['US_at_current_prices_Footnote','US_at_current_prices_MissingValue'])
df_5 = df_5.rename(columns = {'US_at_current_prices_in_millions_Value':'Import_$USMillion'})
df_5
```

```python
# dropping & renaming columns
df_6 = df_6.drop(columns = ['Percentage_of_total_food_Footnote','Percentage_of_total_food_MissingValue'])
df_6 = df_6.rename(columns = {'Percentage_of_total_food_Value':'Import_%TotalFood'})
df_6
```

```python
# dropping & renaming columns
df_7 = df_7.drop(columns = ['Percentage_of_total_merchandise_trade_Footnote','Percentage_of_total_merchandise_trade_MissingValue'])
df_7 = df_7.rename(columns = {'Percentage_of_total_merchandise_trade_Value':'Import_%TotalMerchandise'})
df_7
```

```python
# dropping & renaming columns
df_8 = df_8.drop(columns = ['Growth_rate_yearonyear_Footnote','Growth_rate_yearonyear_MissingValue'])
df_8 = df_8.rename(columns = {'Growth_rate_yearonyear_Value':'Import_YoYGR'})
df_8
```

```python
#merge export datasets
export_dfs_list = [df_1,df_2,df_3,df_4]
df_exports = export_dfs_list[0]

for i in range(1, len(export_dfs_list)):
    df_exports = pd.merge(
        df_exports,
        export_dfs_list[i],
        on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
    )
df_exports
```

```python
#merge import datasets
import_dfs_list = [df_5, df_6,df_7, df_8]
df_imports = import_dfs_list[0]

for i in range(1, len(import_dfs_list)):
    df_imports = pd.merge(
        df_imports,
        import_dfs_list[i],
        on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
    )
df_imports
```

```python
#merge export & import datasets
df_combined = pd.merge(
    df_imports,
    df_exports,
    on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
)
df_combined
```

```python
# Convert Import and Export columns from 'Millions' to absolute $US values.
df_combined['Import_$USMillion'] = df_combined['Import_$USMillion'] * 1000000
df_combined['Export_$USMillion'] = df_combined['Export_$USMillion'] * 1000000

df_combined = df_combined.rename(columns = {"Import_$USMillion":"Import_$US","Export_$USMillion":"Export_$US" })
df_combined
```

```python
#list unique enteries in ProcessFoodCategory_Label column
list(df_combined['ProcessFoodCategory_Label'].unique())
```

```python
#extract aggregage data for all products
df_combined_AllProductsTotal = df_combined[df_combined['ProcessFoodCategory_Label'] == 'All products']
df_combined_AllProductsTotal
```

```python
# extract data for food products broken down by category
product_list = ['Vegetable', 'Fruit', 'Cereals', 'Meat', 'Poultry','Fish','Shellfish','Dairy','Egg','Nuts', 'Oilseeds','Animal fats','Vegetable fats','Non-alcoholic','Alcoholic beverages','Sugars, jams','Cocoa et cie','Edible preparations', 'Other, n.e.c.']

df_combined_FoodProducts = df_combined[df_combined['ProcessFoodCategory_Label'].isin(product_list)]

df_combined_FoodProducts
```

```python
# extract overall aggregate data for food product trade
df_combined_FoodProductsTotal = df_combined[df_combined['ProcessFoodCategory_Label'] == 'Total']
df_combined_FoodProductsTotal
```

```python
# extract foodtrade data broken down by food product category and process
exclude_list = product_list+ ['All products','Total']
df_combined_FoodProducts_Process = df_combined[~df_combined['ProcessFoodCategory_Label'].isin(exclude_list)]
df_combined_FoodProducts_Process[['Process', 'FoodCategory']] = df_combined_FoodProducts_Process['ProcessFoodCategory_Label'].str.split(' - ', expand = True)
df_combined_FoodProducts_Process
```

```python
# factchecking or correlating totals from different datasets
Total_1996_from_df_combined_FoodProducts_Process =  df_combined_FoodProducts_Process[df_combined_FoodProducts_Process['Year'] == 1996]['Import_$US'].sum()
Total_1996_from_df_combined_FoodProducts = df_combined_FoodProducts[df_combined_FoodProducts['Year'] == 1996]['Import_$US'].sum()
Total_from_df_combined_FoodProductsTotal = df_combined_FoodProductsTotal[df_combined_FoodProductsTotal['Year']==1996]['Import_$US']


print(f"Total_1996_from_df_combined_FoodProducts_Process:  {Total_1996_from_df_combined_FoodProducts_Process/1000000} Million")
print(f"Total_1996_from_df_combined_FoodProducts:  {Total_1996_from_df_combined_FoodProducts/1000000} Million")
print(f"Total_from_df_combined_FoodProductsTotal:  {Total_from_df_combined_FoodProductsTotal.item()/1000000} Million")
```

### Data Analysis and Visualization

```python
# Yearly food export and import 
billion_format = lambda x: f'${x/1000000000:.3f}B'
df_combined_FoodProductsTotal[['Year', 'Export_$US','Import_$US']].style.format({'Export_$US': billion_format,
    'Import_$US': billion_format}).hide(axis = "index")
```

```python
# Yearly food export and import trend visual
def billions_formatter(x, pos):
    return '${:1.0f}B'.format(x / 1000000000)
formatter = FuncFormatter(billions_formatter)

fig, ax = plt.subplots(3, 1, figsize=(15, 15))
df_combined_FoodProductsTotal[['Year', 'Import_$US']].plot(
    x='Year',
    y='Import_$US',
    kind='bar',
    ax=ax[0],              
    color='grey',
    width=0.75
)
ax[0].set_title('Nigeria Yearly Food Import Trend ($US)')
ax[0].set_ylabel('Import Amount ($US)')
ax[0].yaxis.set_major_formatter(formatter)

df_combined_FoodProductsTotal[['Year', 'Export_$US']].plot(
    x='Year',
    y='Export_$US',
    kind='bar',
    ax=ax[1],              
    color='green',
    width=0.75
)
ax[1].set_title('Nigeria Yearly Food Export Trend ($US)')
ax[1].set_ylabel('Export Amount ($US)')
ax[1].yaxis.set_major_formatter(formatter)

df_combined_FoodProductsTotal[['Year','Import_$US', 'Export_$US']].plot(
    x='Year',
    y=['Import_$US','Export_$US'],
    kind='bar',
    ax=ax[2],              
    color=['grey','green'],
    width=0.75
)
ax[2].set_title('Nigeria Yearly Food Import & Export Trend ($US)')
ax[2].set_ylabel('Monetary Amount ($US)')
ax[2].yaxis.set_major_formatter(formatter)

plt.tight_layout()

plt.show()
```
