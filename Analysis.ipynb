```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_1 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProc_Export_$USMillions.csv")
df_2 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExport_%totalfood.csv")
df_3 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExport_%totalMerchandise.csv")
df_4 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcExportYoYGR.csv")
df_5 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport$USMilliion.csv")
df_6 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport%TotalFood.csv")
df_7 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcImport%TotalMerchandise.csv")
df_8 = pd.read_csv(r"C:\Users\PC\Desktop\Oluwatobi\Python\International Food Trade\TradeFoodCatByProcimportYoYGR.csv")
```

```python
df_1 = df_1.drop(columns = ['US_at_current_prices_Footnote','US_at_current_prices_MissingValue' ])
df_1 = df_1.rename(columns = {'US_at_current_prices_in_millions_Value':'Export_$USMillion'})
df_1
```

```python
df_2 = df_2.drop(columns = ['Percentage_of_total_food_Footnote','Percentage_of_total_food_MissingValue'])
df_2 = df_2.rename(columns = {'Percentage_of_total_food_Value':'Export_%TotalFood'})
df_2
```

```python
df_3 = df_3.drop(columns = ['Percentage_of_total_merchandise_trade_Footnote','Percentage_of_total_merchandise_trade_MissingValue'])
df_3 = df_3.rename(columns = {'Percentage_of_total_merchandise_trade_Value':'Export_%TotalMerchandise'})
df_3
```

```python
df_4 = df_4.drop(columns = ['Growth_rate_yearonyear_Footnote','Growth_rate_yearonyear_MissingValue'])
df_4 = df_4.rename(columns = {'Growth_rate_yearonyear_Value':'Export_YoYGR'})
df_4
```

```python
df_5 = df_5.drop(columns = ['US_at_current_prices_Footnote','US_at_current_prices_MissingValue'])
df_5 = df_5.rename(columns = {'US_at_current_prices_in_millions_Value':'Import_$USMillion'})
df_5
```

```python
df_6 = df_6.drop(columns = ['Percentage_of_total_food_Footnote','Percentage_of_total_food_MissingValue'])
df_6 = df_6.rename(columns = {'Percentage_of_total_food_Value':'Import_%TotalFood'})
df_6
```

```python
df_7 = df_7.drop(columns = ['Percentage_of_total_merchandise_trade_Footnote','Percentage_of_total_merchandise_trade_MissingValue'])
df_7 = df_7.rename(columns = {'Percentage_of_total_merchandise_trade_Value':'Import_%TotalMerchandise'})
df_7
```

```python
df_8 = df_8.drop(columns = ['Growth_rate_yearonyear_Footnote','Growth_rate_yearonyear_MissingValue'])
df_8 = df_8.rename(columns = {'Growth_rate_yearonyear_Value':'Import_YoYGR'})
df_8
```

```python
export_dfs_list = [df_1,df_2,df_3,df_4]
df_exports = export_dfs_list[0]

for i in range(1, len(export_dfs_list)):
    df_exports = pd.merge(
        df_exports,
        export_dfs_list[i],
        on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
    )
df_exports
```

```python
import_dfs_list = [df_5, df_6,df_7, df_8]
df_imports = import_dfs_list[0]

for i in range(1, len(import_dfs_list)):
    df_imports = pd.merge(
        df_imports,
        import_dfs_list[i],
        on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
    )
df_imports
```

```python
df_combined = pd.merge(
    df_imports,
    df_exports,
    on = ['ProcessFoodCategory_Label','Year'],
        how = 'outer'
)
df_combined
```

```python
df_combined['Import_$USMillion'] = df_combined['Import_$USMillion'] * 1000000
df_combined['Export_$USMillion'] = df_combined['Export_$USMillion'] * 1000000
df_combined
```

```python
df_combined_AllProductsTotal = df_combined[df_combined['ProcessFoodCategory_Label'] == 'All products']
df_combined_AllProductsTotal
```

```python
product_list = ['Vegetable', 'Fruit', 'Cereals', 'Meat', 'Poultry','Fish','Shellfish','Dairy','Egg','Nuts', 'Oilseeds','Animal fats','Vegetable fats','Non-alcoholic','Alcoholic beverages','Sugars, jams','Cocoa et cie','Edible preparations', 'Other, n.e.c.']

df_combined_FoodProducts = df_combined[df_combined['ProcessFoodCategory_Label'].isin(product_list)]

df_combined_FoodProducts
```

```python
df_combined_FoodProductsTotal = df_combined[df_combined['ProcessFoodCategory_Label'] == 'Total']
df_combined_FoodProductsTotal
```

```python
round(df_combined_FoodProducts[df_combined_FoodProducts['Year'] == 1996]['Import_$USMillion'].sum(),-13) == round(df_combined_FoodProductsTotal[df_combined_FoodProductsTotal['Year']==1996]['Import_$USMillion'],-13)
```

```python
exclude_list = product_list+ ['All products','Total']
exclude_list
```

```python
df_combined_FoodProducts_Process = df_combined[~df_combined['ProcessFoodCategory_Label'].isin(exclude_list)]
df_combined_FoodProducts_Process
```

```python
df_combined_FoodProducts_Process[['Process', 'FoodCategory']] = df_combined_FoodProducts_Process['ProcessFoodCategory_Label'].str.split(' - ', expand = True)
df_combined_FoodProducts_Process
```

```python
Total_1996_from_df_combined_FoodProducts_Process =  df_combined_FoodProducts_Process[df_combined_FoodProducts_Process['Year'] == 1996]['Import_$USMillion'].sum()
Total_1996_from_df_combined_FoodProducts = df_combined_FoodProducts[df_combined_FoodProducts['Year'] == 1996]['Import_$USMillion'].sum()
Total_from_df_combined_FoodProductsTotal = df_combined_FoodProductsTotal[df_combined_FoodProductsTotal['Year']==1996]['Import_$USMillion']


print(f"Total_1996_from_df_combined_FoodProducts_Process:  {Total_1996_from_df_combined_FoodProducts_Process/1000000000000} Trillion")
print(f"Total_1996_from_df_combined_FoodProducts:  {Total_1996_from_df_combined_FoodProducts/1000000000000} Trillion")
print(f"Total_from_df_combined_FoodProductsTotal:  {Total_from_df_combined_FoodProductsTotal.item()/1000000000000} Trillion")
```
